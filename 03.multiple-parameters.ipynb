```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    result: str
```

```python
def hello(state: AgentState) ->AgentState:
    state["result"] = "Hello, how are you " + state["name"]
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("say_hello", hello)
graph.set_entry_point("say_hello")
graph.set_finish_point("say_hello")

agent = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(agent.get_graph().draw_mermaid_png()))
```

```python
final_state = agent.invoke({"name": "ali"})
```
