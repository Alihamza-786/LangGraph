# Core Concepts

## Dictionary
A dictionary (in Python and programming) is a data structure that stores data in key–value pairs.

```python
student = {
    "name" : "ali",
    "age": 40
}
```

```python
print(student)
```

## Typed Dictionary
- TypedDict allows you to define your own structure (blueprint) for dictionaries.

- It ensures correct keys + correct data types.

- It makes code safer and easier to read.

```python
from typing import TypedDict
class Student(TypedDict):
    name: str
    age: int
```

```python
Student(name = "ali", age = 34)
```

## Union
Union lets a variable or function accept more than one type.

```python
from typing import Union
def addition(x: Union[int, float]):
    return x + 2
```

```python
addition(3)
```

```python
addition(3.4)
```

## Optional
- Optional is a shortcut for Union[type, None]
- it means a value can be the given type OR None.

Think of Optional like a maybe box:
- Sometimes you have a value.
- Sometimes you don’t (it’s empty, i.e., None).

```python
from typing import Optional

def greet(name: Optional[int]): # Union [str, None]
    if name is None:
        print("Hello, guest!")
    else:
        print(f"Hello, {name}!")
```

```python
greet("Ali")
```

```python
greet(None)
```

## Any
The variable can be any type at all — string, int, float, list, dict, etc.

Any = “I don’t care what type this is, anything goes.”

```python
def print_name(name: Any):
    print(name)
```
