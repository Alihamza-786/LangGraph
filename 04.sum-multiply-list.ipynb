```python
from typing import TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    number: List[int]
    result: str
```

```python
# def calculation(state: AgentState) ->AgentState:
#     sum = 0
#     for num in state["number"]:
#         sum = sum + num

#     state["result"] = f"Hello {state["name"]}, your sum = {sum}"
#     return state
```

```python
def calculation(state: AgentState) ->AgentState:
    multiplication = 1
    for num in state["number"]:
        multiplication = multiplication * num

    state["result"] = f"Hello {state["name"]}, your multiplication = {multiplication}"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("calculation", calculation)
graph.set_entry_point("calculation")
graph.set_finish_point("calculation")

agent = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(agent.get_graph().draw_mermaid_png()))
```

```python
final_state = agent.invoke({"name": "ali", "number": [1,2,3,4,5]})
```
